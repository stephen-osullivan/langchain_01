```python
## Data Ingestion: steps are load -> split -> store
from langchain_community.document_loaders import TextLoader
loader = TextLoader('paradise_lost.txt')
text_documents = loader.load()
text_documents
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
# now we split
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(text_documents)
print('Num of splits:', len(splits))
splits
```

```python
import bs4
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("post-content", "post-title", "post-header") # this classes are found by inspecting the html
        )
    ),
)
web_documents = loader.load()
web_documents
```
